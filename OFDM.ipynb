```python
OFDM
- 고속의 송신 데이터열을, 수백개의 이상의 직교하는 subcarrier로 변조시켜 다중화하는 방식

subcarrier
- 정보를 단일 반송파(single carrier)에만 실어서 보내지 않고, 직교성이 있는 여러 subcarrier에 실어서 전송할 때, 각각 쪼개진 반송파들을 지칭하는 말.
- 각 서브캐리어는 서로 다른 중심 주파수를 가지고 있습니다.
- 각 서브캐리어는 서로 직교 관계에 있기 때문에, 서로 간섭하지 않고 독립적으로 데이터를 전송할 수 있습니다.

- 두 신호가 서로 직교 관계에 있다면, 그들이 상관관계(correlation) 값은 0 입니다.
```

```python
특징
1. subcarrier간 직교성 유지
2. subcarrier 파형
- 시간축) 서로 간섭을 일으키지 않게 직교
- 주파수) 약간 겹치게 됨
3. 주파수 선택적 페이딩이 완화되는 효과를 볼 수 있음.
- 주파수 선택적 페이딩: 특정 주파수 대역의 신호 강도가 감소하는 현상, 다중 경로 전파에 의해 발생.

- 주파수 다이버시티 효과가 기대됨
-> 데이터를 여러 주파수 채널에 분산시켜 전송함으로써, 특정 주파수 채널에서 발생하는 간섭이나 fading으로 인한 성능 저하를 최소화하는 것이 목표.
```
