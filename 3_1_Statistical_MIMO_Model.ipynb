**MIMO channel**

```python
MIMO 시스템: 다중 입출력 시스템(Multiple input Multiple output)
- 무선 통신에서 송수신기에 여러 개의 안테나를 사용하는 기술.
- 데이터 전송 속도와 신뢰성을 크게 향상시키기 위해 사용됩니다.
```

```python
유형
1. SISO: Single Input Single Output
- 하나의 송신 안테나와 하나의 수신 안테나
2. SIMO
3. MISO
4. MIMO
```

```python
 MIMO 이점
 - 향상된 데이터 속도: 여러 경로를 통해 동시에 데이터 전송이 가능해 데이터 전송 속도가 증가합니다.
 - 개선 된 신호 품질: 다중 경러 전송으로 인한 간섭과 페이딩 문제를 완화시켜 신호 품질을 높입니다.
 - 향상된 용량: 주파수 대역폭을 효율적으로 사용하여 네트워크 용량을 증가시킵니다.
```
